```python
count = 134
lldict7 = dict()

while count > 0:
    fnm = mum_hood.iloc[count,1]
    print(fnm)
    f = urllib.request.urlopen(fnm)
    myfile = f.read()
    soup7 = BeautifulSoup(myfile, 'lxml')
    llgeo = soup7.find("span", attrs={ "class" : "geo"})
    if llgeo == None:
        count = count - 1
        print(count)
        continue
    else:
        ll = llgeo.text.split('; ')
        hood_lat = ll[0]
        print(hood_lat)
        hood_long = ll[1]
        print(hood_long)
        lldict7[mum_hood.iloc[count,0]] = (hood_lat, hood_long)
        count = count - 1
        print(count)
    
```
